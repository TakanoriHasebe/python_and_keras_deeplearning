### モデルを最大限に活用するために

#### 高度なアーキテクチャパターン

データを正規化してからモデルに供給していた。<br>
Dense, Conv2Dに渡されるデータが正規化されていたとしても、出力がそうなるとは限らない。<br>

その部分を解消するために、<br>
畳み込み層、全結合層の後にBatchNormalization層を適用する。<br>

```python
# 畳み込みの後
conv_model.add(layers.Conv2D(32, 3, activation='relu'))
conv_model.add(layers.BatchNormalization())

# 全結合の後
dense_model.add(layers.Dense(32, activation='relu'))
dense_model.add(layers.BatchNormalization())
```

#### dw畳み込み

dw畳み込み層は、入力の各チャネルで空間畳み込み演算を別々に実行した後、<br>
pw畳み込み(1×1畳み込み)演算を通じて出力チャネルを連結する。<br>

この処理が意味を持つのは、<br>
「入力の空間的な位置同士は高い相関関係にあるものの、異なるチャネル同士はほぼ独立している」<br>
と想定される場合である。<br>

```python
# 画像の表示
from IPython.display import Image
Image(filename='08.png', width=600)
```

小さなデータセットに基づいて、画像分類タスクのための軽量なdw畳み込みネットワークを構築。<br>

```python
from keras.models import Sequential, Model
from keras import layers
```

```python
height = 64
width = 64
channels = 3
num_classes = 10

model = Sequential()
model.add(layers.SeparableConv2D(32, 3, activation='relu', input_shape=(height, width, channels,)))
model.add(layers.SeparableConv2D(64, 3, activation='relu'))
model.add(layers.MaxPooling2D(2))

model.add(layers.SeparableConv2D(64, 3, activation='relu', input_shape=(height, width, channels,)))
model.add(layers.SeparableConv2D(128, 3, activation='relu'))
model.add(layers.MaxPooling2D(2))

model.add(layers.SeparableConv2D(64, 3, activation='relu', input_shape=(height, width, channels,)))
model.add(layers.SeparableConv2D(128, 3, activation='relu'))
model.add(layers.GlobalAveragePooling2D())

model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(num_classes, activation='softmax'))

model.compile(optimizer='rmsprop', loss='categorical_crossentropy')
```

<ol>
    <li>ハイパーパラメータの集まりを自動的に選択する。</li>
    <li>対応するモデルを構築する。</li>
    <li>モデルに訓練データを学習させ、検証データで最終的な性能を判断する。</li>
    <li>次に試すハイパーパラメータの集まりを（自動的に）選択する。</li>
    <li>手順2~3を繰り返す。</li>
    <li>最後に、テストデータでモデルの性能を評価する。</li>
</ol>
