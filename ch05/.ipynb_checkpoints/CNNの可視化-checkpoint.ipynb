### 中間層の出力を可視化する

```python
from keras.models import load_model
model = load_model('cats_and_dogs_small_2.h5')
model.summary()
```

```python
# 取ってくる画像
img_path = "/Users/Takanori/Downloads/cats_and_dogs_small/test/cats/cat.1700.jpg"
```

```python
# この画像を4次元テンソルとして前処理
from keras.preprocessing import image
import numpy as np
```

```python
img = image.load_img(img_path, target_size=(150, 150)) # イメージを(150, 150)に変換
type(img)
```

```python
img
```

```python
# imageからarrayに変換
img_tensor = image.img_to_array(img)
```

```python
img_tensor.shape
```

```python
# 列方向に次元を１つ広げる
img_tensor = np.expand_dims(img_tensor, axis=0)
```

```python
img_tensor.shape
```

```python
# 以下の方法で前処理されている
img_tensor /= 255.
```

```python
print(img_tensor.shape)
```

```python
from matplotlib import pyplot as plt

plt.imshow(img_tensor[0])
plt.show()
```
