## 小さなデータセットでCNNを一から訓練

犬と猫の4,000枚の画像（2,000枚の犬の画像と2,000枚の猫の画像）が含まれたデータセットで<br>
画像を犬と猫に分類してみる<br>
訓練に2,000枚、検証に1,000枚、テストに1,000枚の画像を使用する。<br>

1. 2,000個の訓練サンプルを使って小さなCNNを単純に訓練する。<br>
これで71%の分類正解率が得られる。<br>

2. データ拡張を行う。<br>
データ拡張により、CNNの正解率は82%に改善される。<br>

次節では、小さなデータセットにディープラーニングを適用するための基本的な手法をさらに２つ紹介。<br>
1. 学習済みのモデルによる特徴抽出。<br>
2. 学習済みのモデルのファインチューニング。<br>

### 小さなデータセットとディープラーニング

モデルが小さく、よく正則化されていて、タスクが単純な場合、数百個で大丈夫。<br>

さらに、ディープラーニングのモデルはそもそも別の目的に使用できる可能性が非常に高い。<br>
例えば、大規模なデータセットで訓練した画像分類モデルや音声テキスト変換モデルをほんの少し変更するだけで<br>
大きく異なる問題に再利用することが可能。<br>

## ネットワークの構築

```python
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
from keras import optimizers

model.compile(loss = 'binary_crossentropy',
              optimizer = optimizers.RMSprop(lr=1e-4),
              metrics=['acc']
             )
```

## データの前処理

現時点ではデータはJPEGファイルとしてハードディスクに格納されている。<br>

CNNに渡す手順は以下のよう。<br>

<ol>
<li>画像ファイルの読み込み</li>
<li>JPEGファイルの内容をRGBのピクセルグリッドにデコードする。</li>
<li>これらのピクセルを浮動小数点型のテンソルに変換</li>
<li>ピクセル値(0 ~ 255)の尺度を取り出し、[0, 1]の範囲の値にする。</li>
</ol>

手順４を行う理由は、NNでは小さな入力値を扱うことが望ましいため。<br>

kerasには上記の手順を自動的に処理するモジュールが含まれている。<br>

#### ImageDataGeneratorを使って、ディレクトリから画像を読み込み

```python
from keras.preprocessing.image import ImageDataGenerator

# すべての画像を1/255でスケーリング
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir, # ターゲットディレクリ
    target_size = (150, 150), # すべての画像サイズを150×150に変更
    batch_size = 20, # バッチサイズ
    class_mode = 'binary' # binary_crossentropyを使用するため、2値のラベルが必要
)

# trainと同様
validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size = (150, 150),
    batch_size = 20,
    class_mode = 'binary'
)
```

```python
for data_batch, labels_batch in train_generator:
    print('data batch shape:', data_batch.shape)
    print('labels batch shape:', labels_batch.shape)
    break
```

## モデルの訓練

train_generator, validation_generatorを用いてモデルをデータに適合する。<br>
これには、fit_generatorメソッドを使用する。<br>

```python
history = model.fit_generator(train_generator, 
                              steps_per_epoch = 100, # バッチサイズが20であるため、すべて終わらせるには100バッチ必要
                              epochs = 30, # 30エポック
                              validation_data = validation_generator, 
                              validation_steps = 50 # バッチサイズが20であるため、すべて終わらせるには50バッチ必要
)
```

```python
model.save('cats_and_dogs_small_1.h5')
```

### 訓練時の損失値と正解率をプロット

```python
import matplotlib.pyplot as plt
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
```

## データ拡張

データ拡張は、既存の訓練サンプルからさらに訓練データを生成するというアプローチを取る。<br>
ランダムな変換をいくつかして、いかにもそれっぽく見せることにする。<br>

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

実際にみてみる

```python
# 画像処理ユーティリティのモジュール
from keras.preprocessing import image

fnames = [os.path.join(train_cats_dir, fname) for fname in os.listdir(train_cats_dir)]

# 水増しする画像を選択
img_path = fnames[3]

# 画像を読み込み、サイズを変更
img = image.load_img(img_path, target_size=(150, 150))

# 形状が(150, 150, 3)のNumpy配列に変換
x = image.img_to_array(img)

# (1, 150, 150, 3)に変形
x = x.reshape((1, ) + x.shape)

# ランダムに変換した画像のバッチを生成
# 無限ループとなるため、何かしらのタイミングでbreakする
i = 0
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i += 1
    if i % 4 == 0:
        break
        
plt.show()
```

データ拡張設定を用いて、新しいCNNを訓練する場合、入力の相関性の高さが問題になる。<br>
ここで過学習を抑制するために、全結合分類器の直前にDropout層を追加する。<br>

```python
# モデルの定義
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150,150,3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
             optimizer=optimizers.RMSprop(lr=1e-4),
             metrics=['acc']
             )
```

```python
# データ拡張ジェネレータを使ってCNNを訓練

# train_datagenでは画像を水増し
train_datagen = ImageDataGenerator(
    rescale = 1. / 255,
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)

# 検証データは水増しすべきでないことに注意
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        train_dir, #　ターゲットディレクトリ
        target_size=(150, 150), # 150×150に変換
        batch_size=32, #  バッチサイズ
        class_mode='binary' # 2値のラベルが必要
)

validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size = (150, 150),
    batch_size = 32,
    class_mode = 'binary'
)

history = model.fit_generator(
    train_generator,
    steps_per_epoch=100,
    epochs=100,
    validation_data = validation_generator,
    validation_steps=50
)
```

```python
model.save('cats_and_dogs_small_2.h5')
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

データ拡張とDropoutのおかげで、過学習が減っている。<br>
